while loop is useful when we want to start/stop executing some codes only when some condition(s) is/are fulfilled. eg. count from 1 to 10 - here the condition is to start from int 1 and keep executing untill the counter does not reach 10.

```python
#1st pattern
i=int(5)
while (i>0):
    j=int(0)
    while(j<i):
        print(j+1,end='')
        j=j+1
    i=i-1
    print('')
```

```python
#2nd pattern
i=int(5)
count=int(1)
while (i>0):
    j=i-count
    while(j<i):
        print(i-j,end='')
        j=j+1
    i=i-1
    count=count+1
    print('')
```

```python
#3rd pattern
i=int(5)
while(i>0):
    j=i
    while(j>0):
        print(j,end='')
        j=j-1
    print('')
    i=i-1
```
